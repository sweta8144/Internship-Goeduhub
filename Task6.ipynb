1.Load the necessary package for plotting using pyplot from matplotlib.

Example - Days(x-axis) represents 8 days and Speed represents a car’s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.

 Days=[1,2,3,4,5,6,7,8]

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Days = np.array([1,2,3,4,5,6,7,8])
Speed = np.array([60,62,61,58,56,57,46,63])
```

3.Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days.

days=[1,2,3,4,5,6,7,8]

```python
days=np.array([1,2,3,4,5,6,7,8])
max_speed=np.array([80,91,92,88,77,79,76,75])
min_speed=np.array([42,43,40,42,33,36,34,35])
avg_speed=np.array([46,58,57,56,40,42,41,36])
```

```python
fig, (axis1, axis2, axis3) = plt.subplots(nrows=3,ncols=1,figsize=(15,15))

axis1.plot(days,max_speed,linewidth=3,linestyle=':')
axis1.scatter(days,max_speed,marker='x',s=150,c='black')
axis1.set_xlabel('Days')
axis1.set_ylabel('Max speed')
axis1.set_title('Days vs Max speed')

axis2.plot(days,min_speed,linewidth=3,linestyle=':')
axis2.scatter(days,min_speed,marker='+',s=150,c='black')
axis2.set_xlabel('Days')
axis2.set_ylabel('Min speed')
axis2.set_title('Days vs Min speed')

axis3.plot(days,avg_speed,linewidth=3,linestyle=':')
axis3.scatter(days,avg_speed,marker='x',s=150,c='black')
axis3.set_xlabel('Days')
axis3.set_ylabel('Avg speed')
axis3.set_title('Days vs Avg speed')
```


4.Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
x = np.linspace(0,5,200)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(15,7))
plt.plot(x,y,':b',linewidth=5)
plt.plot(x,z,'--r',linewidth=5)
plt.xlabel('Angle in radians')
plt.ylabel('Magnitude')
plt.title('Sine/Cosine function curve')
plt.legend(['sine','cosine'])
plt.show()
```


5.Plot Simple bar chart showing popularity of Programming Languages.

Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]

Security = [44 ,36 ,55, 50, 42]

Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = np.array([56,39,34,34,29])
Security = np.array([44,36,55,50,42])
```

```python
gridsize = (2,4)
fig = plt.figure(figsize=(15,7))
ax1 = plt.subplot2grid(gridsize,(0,0), colspan=2, rowspan=1)
ax2 = plt.subplot2grid(gridsize,(1,0), colspan=2, rowspan=41)
ax3 = plt.subplot2grid(gridsize,(0,2), colspan=40, rowspan=40)

ax1.bar(Languages,Popularity,color='green')
ax1.set_ylabel('Popularity')
ax1.set_title('Barplot of popularity')

ax2.bar(Languages,Security,color='red')
ax2.set_ylabel('Security')
ax2.set_title('Barplot ofsecurity')

ax3.hist(Popularity)
ax3.set_xlabel('Bins')
ax3.set_ylabel('Popularity')
ax3.set_title('Histogram of popularity')

fig.suptitle('Programing Languages')
```


6.Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students by marks range and try to figure out how many Students are average, below-average and Excellent.

Marks = [ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]

Histogram showing Below Average, Average and Execellent distribution

40-60: Below Average

60-80: Average

80-100: Excellent

```python
Marks = np.array([61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])
below_average = Marks[np.logical_and(Marks>=40, Marks<60)]
average = Marks[np.logical_and(Marks>=60, Marks<80)]
excellent = Marks[np.logical_and(Marks>=80, Marks<=100)]
```

```python
plt.figure(figsize=(15,7))
plt.hist(Marks, align='mid', color='blue', edgecolor='black')
plt.xlabel('Bins')
plt.ylabel('Marks')
plt.title('Histogram of Marks')
plt.vlines([40, 60, 80,100], ymin=0, ymax=6, linestyles='dashed')
plt.show()
```

```python
print('Number of below average students: ',(below_average.size))
print('Number of average students: ',(average.size))
print('Number of excellent students: ',(excellent.size))
```

```python
plt.figure(figsize=(15,7))
plt.hist(Marks,bins=[40, 60, 80,100], align='mid', color='yellow', edgecolor='black')
plt.xlabel('Bins')
plt.ylabel('Marks')
plt.title('Histogram of Marks')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(9,9))
df.sex.value_counts().plot.pie(labels=['Male','Female'],startangle=180)
plt.title('Male/Female proportion')
```

```python
df_male = df[df['sex']=='male']
df_male.sex.value_counts()
```

```python
df_female = df[df['sex']=='female']
df_female.sex.value_counts()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(df_male.fare,df_male.age,label='Male',color='blue',marker='x')
plt.scatter(df_female.fare,df_female.age,label='Female',color='red')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Scatter plot')
plt.legend()
plt.show()
```
